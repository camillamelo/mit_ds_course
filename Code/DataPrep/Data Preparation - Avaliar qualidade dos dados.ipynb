```python
import pandas as pd

pd.set_option("display.precision", 4)

df_wine = pd.read_parquet("../../Data/Processed/wine/df_wine.parquet") 

if (df_wine.isnull().values.any()): 
    print("Existem dados faltantes na tabela de vinhos") 
else: 
    print("Sem dados faltantes para a tabela de vinhos")
```

```python
''' Visão geral dos dados'''

print('Data shape:', df_wine.shape)
print('Data columns:', df_wine.columns.tolist())
df_wine.head()
```

```python
''' Estatísticas Gerais'''

df_wine.describe()
```

```python
# !Dúvida em relação a quantificação. Numerador e Denominador!


''' Quantificação por variável (%)'''

for (column, _ ) in df_wine.drop(['type'], axis = 1).items():
    print(f'{column} \t Percentual do total: \t {"{:.2f}".format(df_wine[column].count())}')
    
```

```python
''' Avaliar a faixa dimâmica por variável'''

''' Loop por colunas do df retirando a coluna type que é do tipo string formatando o resultado em 2 casa decimais'''

for (column, _ ) in df_wine.drop(['type'], axis = 1).items():
    print(f'{column} \t faixa dinâmica: \t {"{:.2f}".format(df_wine[column].max() - df_wine[column].min())}')
```
