```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df_red = pd.read_csv('../Data/Raw/wine/winequality-red.csv', sep=';')
df_red
```

```python
df_white = pd.read_csv('../Data/Raw/wine/winequality-white.csv', sep=';')
df_white
```
