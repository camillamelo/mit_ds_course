# Projeto Vinhos
## Normalização
### Imports

```python
import pandas as pd
import numpy as np
import seaborn as sbn
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn import metrics
from sklearn import linear_model
```

### Importando dados

```python
dados = pd.read_csv('Data/Processed/retail/bigTableWines.csv')
dadosNumericos = dados.drop('Tipo Vinho', axis=1)
dados.head(5)
```

### Standard Scaling

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler()
scaler.fit(dadosNumericos)
std_normed_data = pd.DataFrame(scaler.transform(dadosNumericos), columns=dadosNumericos.columns)

for coluna in dadosNumericos.columns:
    plt.figure(figsize=[7,5])
    sbn.distplot(dadosNumericos[coluna], label='Dados Originais', )
    sbn.distplot(std_normed_data[coluna], label='Standard Scaling')
    plt.legend()
```

### MIN / MAX

```python
scaler = preprocessing.MinMaxScaler()
scaler.fit(dadosNumericos)
normed_data = pd.DataFrame(scaler.transform(dadosNumericos), columns=dadosNumericos.columns)

for coluna in dadosNumericos.columns:
    plt.figure(figsize=[7,5])
    sbn.distplot(dadosNumericos[coluna], label='Dados Originais')
    sbn.distplot(normed_data[coluna], label='Min/Max')
    plt.legend()
```

### Logarítimo

```python
for coluna in dadosNumericos.columns:
    normed = np.log1p(dadosNumericos[coluna])
    plt.figure(figsize=[7,5])
    sbn.distplot(dadosNumericos[coluna], label='Dados Originais')
    sbn.distplot(normed, label='Norm por Log')
    plt.legend()
```
